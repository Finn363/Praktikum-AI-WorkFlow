```python
import pandas as pd
data = {
    "tanggal":["2025-02-01","2025-02-02","2025-02-03","2025-02-04","2025-02-05","2025-02-06","2025-02-07","2025-02-08","2025-02-09","2025-02-10"],
    "produk":["bawang","pecin","masako","gula","kopi","susu","teh manis","cilok","kurupuk","bensin"],
    "jumlah terjual":[10, 3, 5, 6, 4, 8, 9, 7, 3, 5],
    "stok":[20, 7, 8, 15, 14, 10, 13, 12, 9, 11],
    "harga satuan":[1500, 4000, 6000, 14000, 5000, 7000, 5000, 3000, 12000, 8000],

}

dt = pd.DataFrame(data)
dt.to_csv("Alfian Arsyad Wijaya 2306011 pertemuan 1 latihan workflow AI.csv")
dt
```

```python
dt["tanggal"] = pd.to_datetime(dt["tanggal"])

dt["total penjualan"] = dt["jumlah terjual"] * dt["harga satuan"]
dt["keutungan"] = dt["total penjualan"] * (dt["jumlah terjual"] * 10000)
dt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X = dt[["jumlah terjual", "stok"]]
Y = (dt["stok"] < 5).astype(int)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
print("akurasi model: ", accuracy_score(Y_test, Y_pred))
```

```python
produk_baru = [[8, 3]]
prediksi = model.predict(produk_baru)

if prediksi [0] == 1:
  print("produk perlu direstok!")
else:
  print("stok masih cukup")
```
